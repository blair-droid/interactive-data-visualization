- import packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# load dataset
df = pd.read_csv("credit_cards_clean.csv")
```

- histogram

```python
df.hist(figsize=(10,8), bins=30)
plt.show()
```

- correlation heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.show()
```

- scatter plots

```python
sns.scatterplot(df, x="PURCHASES_TRX", y="PURCHASES")
plt.show()
```

```python
sns.scatterplot(df, x="CREDIT_LIMIT", y="BALANCE")
plt.show()
```

```python
sns.scatterplot(df, x="CASH_ADVANCE_TRX", y="CASH_ADVANCE")
```

- repayment histogram

```python
sns.histplot(df["PRC_FULL_PAYMENT"], bins=20)
plt.show()
```
